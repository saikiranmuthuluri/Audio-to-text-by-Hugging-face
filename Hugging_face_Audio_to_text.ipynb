```python
# Installing required libraries
import librosa
import torch
import IPython.display as display
from transformers import Wav2Vec2ForCTC,Wav2Vec2Tokenizer
import numpy as np
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
# Loading Audio file
audio,sampling_rate=librosa.load("/content/OSR_us_000_0019_8k.wav",sr=16000)
```

```python
audio,sampling_rate
```

```python
# Displaying a audio
display.Audio("/content/OSR_us_000_0019_8k.wav",autoplay=True)
```

```python
input_values = tokenizer(audio,return_tensor = "pt").input_values
```

```python
input_values
```

```python
logits = model(torch.tensor(input_values).type(torch.float32)).logits
```

```python
logits
```

```python
predicted_ids = torch.argmax(logits,dim=-1)
```

```python
# Tokenizing the words
transcriptions = tokenizer.decode(predicted_ids[0])
```
